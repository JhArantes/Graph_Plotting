# Importando e trazendo DFs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fnotas = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/ml-latest-small/ratings.csv")
fnomes = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/ml-latest-small/movies.csv")
fnotas.columns = ['idUsuario', 'idFilme', 'nota', 'momento']
fnomes.columns = ['idFilme', 'nome', 'genero']
display(fnotas.head(3))
display(fnomes.head(3))
```

# Separando e agrupando dados

```python
# Separando e tirando media
fnotas.query('idFilme==1')['nota'].mean()
fnotas.query('idFilme==2')['nota'].mean()

# Medias de todos os filmes
medias_filmes = fnotas.groupby('idFilme')['nota'].mean()
medias_filmes.head(2)
```

# Plotando dados

```python
medias_filmes.plot(kind='hist')
```

```python
sns.histplot(medias_filmes)
```

```python
sns.boxplot(medias_filmes)
```

## Estilizando Grafico


```python
dados_brasil = pd.read_csv("C:\\all_my_dfs\\brasil_imigrantes.csv")

fig, ax = plt.subplots(figsize=(8, 4))                   ######
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw='3')
#                                                              ##################
ax.set_title('Imigrantes Brasil para o Canada\n 1980 a 2013', fontsize=18, loc='left')
ax.set_ylabel('Numero de imigrantes', fontsize=14)
ax.set_xlabel('Anos', fontsize=14)    ############
#                        ##########
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()
# MUDANÇAS
#  FontaSize dos ticks/XY_ticks    ax.XYaxis.set_tick_params(labelsize=12)
#  FontSize das letras e positionfontsize=18, loc='left'      XYlabel, Title 
#  Espesura das linhas  LW lw   
```

imcompleto


```python
fig, axs = plt.subplots(1, 2, figsize=(18, 9))
plt.subplots_adjust(wspace=0.3, hspace=0.4)
                
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])                                                     
axs[0].set_title('Imigrantes Brasil para o Canada\n 1980 a 2013')

axs[1].plot(dados_brasil['ano'], dados_brasil['imigrantes'])                                                     
axs[1].set_title('Imigrantes Brasil para o Canada\n 1980 a 2013')

for ax in axs.flat:
    ax.xaxis.set_tick_params(labelsize=12)
    ax.yaxis.set_tick_params(labelsize=12)
    ax.set_xlabel('Mês', fontsize=14)
    ax.set_ylabel('Número de vendas', fontsize=14)

    

plt.show()
```
